```python
def Entropy(data):
    ct, res = 0., 0.
    for a in data:
        ct+=1
    d = dict((w[-1],0) for w in data)
    for w in data:
        d[w[-1]]+=1.0
    for k,v in d.items():
        p = d[k]/ct
        res -= p*math.log(p)
        d[k]=p
    return res
    
def EntropyByIndex(data , index):
    d = dict((w[index],0) for w in data)
    res = 0
    for w in data:
        d[w[index]]+=1.0
    for k,v in d.items():
        p = d[k]/len(data)
        #kent=p*Entropy([w for w in data if w[index]==k])
        kent = p*EntropyByIndexValue(data, index, k)
        res += kent
        d[k]=kent
    return res

def EntropyByIndexValue(data, index, k):
    ct, res = 0., 0.
    for w in data:
        if w[index]==k:
            ct+=1
    d = dict((w[-1],0) for w in data if w[index]==k)
    for w in data:
        if(w[index]==k):
            d[w[-1]]+=1.0
    for k,v in d.items():
        p = d[k]/ct
        res -= p*math.log(p)
        d[k]=p
    return res


def MinEntropy(data, attr):   #-->(attr[i], entropy)
    d = dict((n,EntropyByIndex(data, n)) for n in attr)
    return min(d, key=d.get)
```

```python
import pandas as pd
import math


class Treenode():
    def __init__(self, maxdepth=2, key=None):
        self.key = key
        self.childs={}       #словарь ключ-потомок
        self.attr = []
        self.data=[]         #данные узла. в root - все
        self.depth=0
        self.maxdepth=maxdepth
        self.splitIndex=0
    def DefaultF(data, attr):
        return attr[0]
       
        
    def SetData(self, data, f =None):
        self.attr= [i for i in range(len(data[0])-1)]
        if not f:
            f = Treenode.DefaultF
        self._addData(data, self.attr, f)
    
    def PredictList(self, attr):
        k=  self.splitIndex
        if k < len(attr) and attr[k] in self.childs:
            return self.childs[attr[k]].PredictList(attr)
        else:
            return [row[-1] for row in self.data]
    def PredictW(self, attr):
        k=self.splitIndex
        if k < len(attr) and attr[k] in self.childs:
            return self.childs[attr[k]].PredictW(attr)
        else:
            d={}
            for w in self.data:
                d.setdefault(w[-1],0)
                d[w[-1]]+=1
            return max(d,key=d.get)
    
    def Predict(self, attr):
        #tt=self._PredictProb(attr)
        #return max(tt, key=tt.get)
        #return self.PredictW(attr)
        l = self.PredictList(attr)
        return max(set(l), key=l.count)
    
    def _addData(self, data, attr, f):
        self.data = data
        if len(attr)!=0 and self.depth!=self.maxdepth:
            self.splitIndex = f(data, attr)
            self.entropy=EntropyByIndex(data,self.splitIndex)
            self.attr = [v for v in attr if self.splitIndex!=v]
            d = SplitByColumn(data, self.splitIndex)
            for k,v in d.items():
                ch = self.childs.setdefault(k,Treenode(self.maxdepth, k))
                ch.depth = self.depth+1
                ch._addData(v, self.attr,f)           
        else:
            self.data = data
    
    def _dataRepr(self):
        prefix = 2*self.depth * "-"
        if len(self.childs)==0:
            return prefix+ ">>" + str(self.data) + "\n"
        res=""
        for k,v in self.childs.items():
            #res += self.depth * " " + f"depth: {self.depth}, key:{str(k)} " + "\n" + self.depth * " " + v._dataRepr() + "\n"
            res+= prefix + f"k[{self.splitIndex}]={k}, entropy={self.entropy}\n" + v._dataRepr()
        return res
    def __repr__(self):
        return self._dataRepr()
    def Count(self):
        res=0
        if(len(self.childs)!=0):
            for k,v in self.childs.items():
                res+=v.Count()
        else:
            if self.data:
                res+=len(self.data)
        return res
    
    
def SplitByColumn(list2d, n):
    d ={}
    for i in range(0, len(list2d)):
        d.setdefault(list2d[i][n], []).append(list2d[i])
    return d

def SplitData(arr, koef1):
    n = int(len(arr)*koef1 + .5)
    return tuple([arr[:n], arr[n:]])
data=[
    [1,0,'cat'],
    [1,0,'cat'],
    [1,0,'cat'],
    [1,0,'cat'],
    [1,3,'dog'],

    [1,0,'dog'],
    [1,0,'dog'],
    [1,0,'dog'],
    [1,0,'dog'],
    [1,0,'cat'],
]


Entropy(data)
EntropyByIndex(data,1)
print(MinEntropy(data,[1,1,0]))
print("----------")



t = Treenode(5, "root")
t.SetData(data, MinEntropy)
print(t)
print(t.Count())

#print("W: ",t.PredictW([1]))
#print(t)
#print(t.PredictList([2,9]))

#print(SplitData(data, .5))

#data = [df.iloc[i][0:20].tolist() for i in range(0,4)]
#print("data:", *data,'\n',sep='\n',)
#tree=DecisionTree(data,20)
#print(tree)
#


#df = pd.read_csv("connect-4.csv")
#firstN = df.iloc[0:4]
#firstN.head()
```

```python
import random
class RandomForest():
    def __init__(self, maxdepth=5, treecount=10, koef = 0.3):
        self.maxdepth=maxdepth
        self.treecount=treecount
        self.koef=koef
        self.trees = [Treenode(self.maxdepth, "root") for i in range(treecount)] 
    def fit(self, data, f=None):
        for t in self.trees:
            t.SetData(random.sample(data, int(self.treecount*self.koef+.5)),MinEntropy)
    def Predict(self, feaut):
        d=dict()
        for t in self.trees:
            lst = t.PredictList(feaut)
            for w in set(lst):
                d.setdefault(w,0)
                d[w]+=lst.count(w)/float(len(lst))
        return max(d, key=d.get)
```

```python
df = pd.read_csv("connect-4.csv")#[]#[50000:58000]
#print(df[:33000].groupby('win').count())
#print(df[33000:].groupby('win').count())

pdata=df.values.tolist()
#print(pdata[:1])


#pddata = [df.iloc[i][0:20].tolist() for i,v in enumerate(df.iterrows())]
depth = 6
train, validate = SplitData(pdata, .50)
tree = Treenode(depth,"root")
tree.SetData(train)
print("tree created")
tree2 = Treenode(depth,"root")
tree2.SetData(train, MinEntropy)
print("tree2 created")
forest = RandomForest(maxdepth=depth, treecount=30, koef=.7)
forest.fit(train,MinEntropy)
print("forest created")

print(tree.Count())
print(tree2.Count())


#print(tree)
#int("\n\n\n\n\n\ntree2-------------===================\n", tree2)
#//print(tree2)

def TestAccuracy(tree, train, validate):
    
    
    truect=0
    for a in validate:
        actual = a[-1]
        predicted = tree.Predict(a[:-1])
        #print(actual,"-->", predicted)
        if(actual==predicted):
            truect+=1
    print(f"Train/validate: {len(train)}/{len(validate)}. Accuracy: {float(truect)/len(validate)}")
    
#print(len(train), *train,sep='\n')

#print(len(validate), *validate,sep='\n')
validate = [a for a in validate if a[-1]=='draw']
TestAccuracy(tree, train, validate)
TestAccuracy(tree2, train, validate)
TestAccuracy(forest, train, validate)
```
